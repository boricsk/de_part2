```python
import pandas as pd
import requests
```

```python
url = r"https://archive-api.open-meteo.com/v1/era5?latitude=52.52&longitude=13.41&start_date=2021-01-01&end_date=2021-12-31&hourly=temperature_2m"
response = requests.get(url)
response.json()
```

```python
url = r"https://archive-api.open-meteo.com/v1/era5"
startDate = '2023-06-01'
params = {
    "latitude" : 41.85,
    "longitude" : -87.65,
    "start_date" : startDate,
    "end_date" : startDate,
    "hourly" : "temperature_2m,wind_speed_10m,precipitation,rain"
}


response = requests.get(url, params=params)
response.json()
```
